## Forward Model Pass

Validate `x_1024 [B, 3, 1024, 1024]` -> `lung_mask [B, 1, 1024, 1024]`.

```python
device = torch.device("cpu")
model = Component4MedSAM(backend="mock").to(device)

model.eval()

# Simulate a batch of 1 scaled 1024x1024 SAM Normalized image
dummy_x_1024 = torch.rand(1, 3, 1024, 1024)

with torch.no_grad():
    lung_mask = model(dummy_x_1024)

print("Input: ", tuple(dummy_x_1024.shape))
print("Lung mask tensor: ", tuple(lung_mask.shape))
```
